```python
import pandas as pd
df = pd.read_csv('/Users/evelynhuang22/projects/salary_prediction/Salary_Data.csv')
```

```python
#remvoe rows where Salary is NaN
df = df[df[['Salary']].notna().all(axis=1)]

```

```python
#CONNECT N CREATE DB

import pyodbc

conn = pyodbc.connect(
    'DRIVER={};' #CHANGE (EG.ODBC Driver 17 for SQL Server)
    'SERVER=;' #CHANGE (EG.localhost)
    'DATABASE=;'  #CHANGE (EG.master)
    'UID=;'  #CHANGE TO OWN UID
    'PWD=', #CHANGE TO OWN PWD
    autocommit=True  
)

cursor = conn.cursor()
cursor.execute("CREATE DATABASE salary_prediction;")
print("DB created")
```

```python
#CREATE TABLE
#GENDER
cursor.execute("""
    CREATE TABLE GENDER (
        GENDER_ID VARCHAR(5) PRIMARY KEY, 
        GENDER VARCHAR(15)
    )
""")

#EDUCATION
cursor.execute("""
    CREATE TABLE EDUCATION (
        EDUCATION_ID VARCHAR(5) PRIMARY KEY, 
        EDUCATION_LEVEL VARCHAR(30)
    )
""")

#JOB TITLE
cursor.execute("""
    CREATE TABLE JOB (
        JOB_ID VARCHAR(10) PRIMARY KEY, 
        JOB_TITLE VARCHAR(100)
    )
""")

print("Tables for gender, education, and job titles are created. ")
```

```python
#CREATE USER TABLE
#USER DETAILS
cursor.execute("""
    CREATE TABLE USERDETAILS (
        USER_ID VARCHAR(10) PRIMARY KEY,
        AGE INT,
        GENDER_ID VARCHAR(5),  
        EDUCATION_ID VARCHAR(5),
        JOB_ID VARCHAR(10),
        YOE DECIMAL(4,2),
        SALARY INT,
        FOREIGN KEY (GENDER_ID) REFERENCES GENDER(GENDER_ID),
        FOREIGN KEY (EDUCATION_ID) REFERENCES EDUCATION(EDUCATION_ID),
        FOREIGN KEY (JOB_ID) REFERENCES JOB(JOB_ID)
    )
""")

print("Table of user detail created. ")
```
